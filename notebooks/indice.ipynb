```python
import json

# Carga el índice desde el archivo
with open('../index.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Determina dónde está la lista de documentos
if isinstance(data, dict):
    # Si el JSON raíz es un dict y la lista está bajo alguna clave, ajústalo aquí:
    docs = data.get('value', data.get('documents', []))
elif isinstance(data, list):
    docs = data
else:
    raise ValueError(f"Estructura inesperada: {type(data)}")

# ID que quieres inspeccionar
target_id = "SpVC1lT8GHgf5dn262OHlZz1UVB_oD_ZdtiN0GbU"

# Busca el documento
doc = next((d for d in docs if isinstance(d, dict) and d.get("id") == target_id), None)

if doc is None:
    print(f"No se encontró ningún documento con id={target_id}")
else:
    # Muestra las claves disponibles
    print("Claves del documento:")
    for key in doc.keys():
        print(" -", key)
    
    # (Opcional) Muestra el documento completo formateado
    print("\nContenido completo:")
    print(json.dumps(doc, indent=2, ensure_ascii=False))
```
