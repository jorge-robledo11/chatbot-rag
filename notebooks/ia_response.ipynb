```python
import json
from IPython.display import display, HTML
import markdown
from pathlib import Path

root_dir = Path.cwd()
json_file_path = root_dir / "json" / "chat_history.json"

# Si no se encuentra, se intenta subiendo un nivel desde el directorio actual.
if not json_file_path.exists():
    root_dir = root_dir.parent
    json_file_path = root_dir / "json" / "chat_history.json"

chat_data = None
try:
    with open(json_file_path, 'r', encoding='utf-8') as f:
        chat_data = json.load(f)
    print(f"‚úÖ Archivo 'chat_history.json' cargado exitosamente")
    
except FileNotFoundError:
    print(f"‚ùå Archivo 'chat_history.json' no encontrado")
```

```python
if chat_data:
    # Bucle para procesar y mostrar la conversaci√≥n con formato.
    for turn in chat_data.get("conversation_history", []):
        user_data = turn.get("user", {})
        ia_data = turn.get("ia", {})

        # --- Formateo del Mensaje del Usuario ---
        user_metadata = (
            f"session: {user_data.get('session_id', 'N/A')} | "
            f"priority: {user_data.get('priority', 'N/A')} | "
            f"type: {user_data.get('query_type', 'N/A')}"
        )
        
        user_html = f"""
        <div style="margin-bottom: 20px; padding: 10px; border-radius: 10px; background-color: #e6ffed;">
            <p style="font-size: 1.2em; margin: 0;">
                <span style="font-size: 1.5em;">üßë</span>
                <strong style="color: #006400;">Usuario:</strong>
            </p>
            <p style="font-size: 1.1em; color: #006400; margin: 5px 0 10px 35px;">
                {user_data.get('query')}
            </p>
            <p style="font-family: monospace; font-size: 0.8em; color: #555; margin: 0 0 0 35px;">
                {user_metadata}
            </p>
        </div>
        """
        display(HTML(user_html))

        # --- Formateo de la Respuesta de la IA ---
        ia_metadata = (
            f"session: {ia_data.get('session_id', 'N/A')} | "
            f"score: {ia_data.get('confidence_score', 'N/A')} | "
            f"escalate: {ia_data.get('escalation_recommended', 'N/A')}"
        )
        
        # Convierte la respuesta Markdown de la IA a HTML
        ia_response_html = markdown.markdown(ia_data.get('response', ''))

        ia_html = f"""
        <div style="margin-bottom: 20px; padding: 10px; border-radius: 10px; background-color: #f3e8ff;">
            <p style="font-size: 1.2em; margin: 0;">
                <span style="font-size: 1.5em;">ü§ñ</span>
                <strong style="color: #5a009d;">IA:</strong>
            </p>
            <div style="color: #5a009d; margin: 5px 0 10px 35px;">
                {ia_response_html}
            </div>
            <p style="font-family: monospace; font-size: 0.8em; color: #555; margin: 0 0 0 35px;">
                {ia_metadata}
            </p>
        </div>
        """
        display(HTML(ia_html))
```
